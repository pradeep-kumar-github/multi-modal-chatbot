## Store Assistant Conversational ChatBot

```python
import os
from openai import OpenAI
import gradio as gr
from IPython.display import Markdown, display
```

```python
MODELllama = 'llama3.2'
MODELPhi = 'phi3'
openai = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')

#if not running, run "ollama serve" at a command line
```

```python
system_message = "You are a helpful assistant"
```

CALLBACK function: the job of this function is to take a new message, take the prior conversation, and return the response

## chat(message, history)

```python
def chat_with_history(message, history):
    history = [{"role" : h["role"], "content" : h["content"]} for h in history]
    messages = [{"role" : "system", "content" : system_message}] + history + [{"role" : "user", "content" : message}]
    stream = openai.chat.completions.create(model = MODELllama, messages = messages, stream = True)

    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
    
```

```python
gr.ChatInterface(fn=chat_with_history).launch()
```

will encourage you to buy but other items if you ask for belt in runtime

```python

def chat_with_history(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    relevant_system_message = system_message
    ## dynamic system message, only if user ask for belt
    if 'belt' in message.lower():
        relevant_system_message += " The store does not sell belts; if you are asked for belts, be sure to point out other items on sale."
    
    messages = [{"role": "system", "content": relevant_system_message}] + history + [{"role": "user", "content": message}]

    stream = openai.chat.completions.create(model = MODELllama, messages = messages, stream = True)

    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn = chat_with_history).launch()
```
